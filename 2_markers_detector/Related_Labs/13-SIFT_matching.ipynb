```python
img1 = cv.imread('roofs1.jpg',cv.IMREAD_GRAYSCALE)
img2 = cv.imread('roofs2.jpg',cv.IMREAD_GRAYSCALE)
```

```python
plt.figure( figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(img1, vmin=0, vmax=255, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img2, vmin=0, vmax=255, cmap="gray")
```

```python
sift_detector = cv.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift_detector.detectAndCompute(img1,None)
kp2, des2 = sift_detector.detectAndCompute(img2,None)
```

```python
img1_features = cv.drawKeypoints( img1, kp1, None, (0,255,0), cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS )
img2_features = cv.drawKeypoints( img2, kp2, None, (0,255,0), cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS )

plt.figure( figsize=(10,10))
plt.imshow(img1_features)
plt.figure( figsize=(10,10))
plt.imshow(img2_features)
```
