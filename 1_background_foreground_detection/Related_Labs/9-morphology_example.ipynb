```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load image and display it

I = cv.imread("galaxy.png", cv.IMREAD_GRAYSCALE) # Convert in grey scale

plt.figure( figsize=(10,10))
plt.imshow( I, cmap="gray")
```

```python
# Our kernel
se = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
print(se)

plt.figure( figsize=(5,5))
plt.imshow(se)
```

```python
opened = cv.morphologyEx(I, cv.MORPH_OPEN, se) # Erosion followed by Dialtation = Opening

plt.figure( figsize=(10,10))
plt.title("Opening: remove all the small stars")
plt.imshow( opened, cmap="gray")
```

```python
tophat = I - opened # the one that have been deleted
# f - its opening

plt.figure( figsize=(10,10))
plt.title("Top-Hat = f - its opening")
plt.imshow( tophat, cmap="gray")
```
