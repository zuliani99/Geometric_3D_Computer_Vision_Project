```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

# Fourier frequencies

```python
# a -> amplitude
# theta -> angle
def complex_from_a_theta(a, theta):
    return a*(np.cos(theta/180.0*np.pi) + 1j*np.sin(theta/180.0*np.pi) )
```

---
Fourier all 0 -> output image all 0

```python
M,N=64,64
S = np.zeros( (M,N), dtype=np.complex64 )
# Matrix of 64x64 complex numbers (all zeros), amplitude and phase all equal to zero

assert np.allclose( np.imag( np.fft.ifft2( S, norm="forward" ) ), 0 )
# Check all close to zero (immaginry part) of the inverse fourier transform of S

# Casted to real number, but id we mantain the propriety of symmetric values are also complex conjugate, the result will be real
I = np.real( np.fft.ifft2( S, norm="forward" ) ) # 2D - Inverse Fourier Transform
plt.figure( figsize=(15,5) )
plt.subplot(1,2,1)
plt.imshow(np.abs(S), interpolation="none")
plt.subplot(1,2,2)
plt.imshow(I)
plt.colorbar()
```

---
- Value of coordinates (1, 0) (y, x) -> control the component of frequency 1 in *y* direction and frequency 0 in *x* direction
- The **amplitude** 1 controls the range of intensity values of the output image, in this case is from -2 to 2
- By setting **angle** 0, the resulting image contains a sinusoid with phase 0
- Since it is the component with frequency q, the sinusoid takes 1 entire image to go through an entire period

```python
M,N=64,64
S = np.zeros( (M,N), dtype=np.complex64 )

S[1,0] = complex_from_a_theta( 1, 0 )
S[-1, 0] = np.conj(S[1,0])
# np.conj -> the complex conjugate of a complex number is obtained by changing the sign of its imaginary part.

assert np.allclose( np.imag( np.fft.ifft2( S, norm="forward" ) ), 0 )

I = np.real( np.fft.ifft2( S, norm="forward" ) )
plt.figure( figsize=(15,5) )
plt.subplot(1,2,1)
plt.imshow(np.abs(S), interpolation="none")
plt.subplot(1,2,2)
plt.imshow(I)
plt.colorbar()
```

---
- Same as before, but now the changed value is the respective component with frequency 0 in the y direction adn frequency 1 in the x direction
- The result is the same as the previous output image, but is roteated of 90 degrees

```python
M,N=64,64
S = np.zeros( (M,N), dtype=np.complex64 )

S[0,1] = complex_from_a_theta( 1, 0 )
S[0,-1] = np.conj(S[0,1])

assert np.allclose( np.imag( np.fft.ifft2( S, norm="forward" ) ), 0 )

I = np.real( np.fft.ifft2( S, norm="forward" ) )
plt.figure( figsize=(15,5) )
plt.subplot(1,2,1)
plt.imshow(np.abs(S), interpolation="none")
plt.subplot(1,2,2)
plt.imshow(I)
plt.colorbar()
```

```python
I = cv.imread("apollo.jpg", cv.IMREAD_GRAYSCALE ) # Convert in grey scale
S = np.fft.fft2(I, norm="ortho") #  2D - DFT


plt.figure( figsize=(15,5) )
plt.subplot(1,3,1)
plt.imshow(I, cmap="gray")
plt.subplot(1,3,2)
#			log magnitude
plt.imshow( np.log(np.abs(S)+1), interpolation="none")
plt.subplot(1,3,3)
#			shifted log magnitude
plt.imshow( np.log(np.abs( np.fft.fftshift(S) )+1), interpolation="none")
```

```python
from ipywidgets import interact, widgets

Sidx = np.argsort( np.abs(S.flatten()) )
    
def keep_frequencies( nfreqs ):
    S2 = np.copy(S)
    S2[ np.unravel_index(Sidx[:-nfreqs],S2.shape) ]=0 # set to zero all the value up to the nfreqs choosen 

    I2 = np.real( np.fft.ifft2( S2, norm="ortho" ) ) # take only the real part and compure the Inverse 2D DFT
    
    plt.figure( figsize=(15,5) )
    plt.subplot(1,3,1)
    plt.imshow(I, cmap="gray", vmin=0, vmax=255)
    plt.subplot(1,3,2)
    plt.imshow(I2, cmap="gray", vmin=0, vmax=255)
    plt.subplot(1,3,3)
    plt.imshow( np.log(np.abs(S2)+1), interpolation="none")

interact( keep_frequencies, nfreqs=(1,10000,2) )
```

# Ideal lowpass filter

```python
from ipywidgets import interact, widgets

    
def lowpass( cutoff ):
    
    S2 = np.copy(S)
    
    for u in range(S2.shape[1]//2+1): # iterate y
        for v in range(S2.shape[0]//2+1): # iterate x
            if np.sqrt( u**2 + v**2 )>cutoff: # those are outside set them to 0
                S2[v,u] = 0
                S2[-v,u] = 0
                S2[v,-u] = 0
                S2[-v,-u] = 0
    
    I2 = np.real( np.fft.ifft2( S2, norm="ortho" ) )
    
    plt.figure( figsize=(20,5) )
    plt.subplot(1,3,1)
    plt.imshow(I, cmap="gray", vmin=0, vmax=255)
    plt.subplot(1,3,2)
    plt.imshow(I2, cmap="gray", vmin=0, vmax=255)
    plt.subplot(1,3,3)
    plt.imshow( np.log(np.abs(S2)+1), interpolation="none")

interact( lowpass, cutoff=(1,100,2) )
```
